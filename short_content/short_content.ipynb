# Case Study: Impact of Top Creators on YouTube and TikTok

In this case study, we analyze the influence of leading content creators across **YouTube** and **TikTok** using a comprehensive dataset of 2025's top creators. We focus on how creator content drives engagement—specifically through **views** and **total engagements**—and how this translates to overall impact on each platform.

## Objectives

- Compare aggregate view counts and engagement rates for creators on YouTube vs. TikTok.
- Identify metrics and patterns associated with viral success (e.g., engagement velocity, share/save rates).
- Analyze audience engagement (likes, dislikes, comments, shares, saves) across platforms.
- Assess how top creators shape platform trends and community behavior.

## Data Analysis Approach

We will explore and analyze:
- Creator-level video counts, total views, and average views per video
- Engagement metrics: likes, dislikes, comments, shares, saves, and total engagements
- Growth indicators: average engagement per 1k views, engagement velocity, follower/subscriber growth (if available)
- Cross-platform strategies and performance comparisons

## Expected Outcomes

By the end of this analysis, we aim to:
- Quantify and compare the impact of top creators on each platform
- Reveal actionable insights into content and engagement strategies that maximize reach
- Provide recommendations for aspiring creators based on observed trends

Let's begin by loading and exploring the dataset!

```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('top_creators_impact_2025.csv')

# Display basic info
display(df.head())
display(df.info())
display(df.describe())

# Check for missing values
display(df.isnull().sum())

# Correlation heatmap
plt.figure(figsize=(12, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()

# Distribution plots for key metrics
numeric_cols = ['n_videos', 'views', 'avg_er', 'likes_sum', 'dislikes_sum', 'comments_sum', 'shares_sum', 'saves_sum', 'total_engagements']
for col in numeric_cols:
    if col in df.columns:
        plt.figure()
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(f'Distribution of {col}')
        plt.show()

# Pairplot for selected features
selected_cols = [col for col in numeric_cols if col in df.columns]
if len(selected_cols) >= 2:
    sns.pairplot(df[selected_cols])
    plt.show()
```

# Exploratory Data Analysis (EDA) Findings

## 1. Correlation Structure
- I noticed two strong **clusters of correlated features**:
  - **Raw counts**: `n_videos`, `views`, `likes_sum`, `comments_sum`, `shares_sum`, `saves_sum`, `total_engagements`.
  - **Engagement ratios**: `avg_comment_ratio`, `avg_share_rate`, `avg_save_rate`, `avg_like_dislike_ratio`, `avg_engagement_per_1k`, `avg_like_rate`.
- Within each cluster, correlations are extremely high (often above 0.9), which points to **multicollinearity**.
- `avg_dislike_rate` stands out for its strong negative correlation with the engagement ratio features.
- `creator_avg_views` looks fairly independent, except for a moderate correlation with `avg_engagement_velocity`.

## 2. Distributions
- Most of the **count-based features** are **heavily right-skewed** with long tails.
- `avg_er` (engagement rate) looks **bimodal**, which could indicate two distinct groups of creators.
- I also spotted several **extreme outliers**, especially in `views`, `likes_sum`, and other totals.

## 3. Scatter Relationships
- Engagement totals are **almost perfectly linear** with each other, confirming redundancy.
- There’s a clear **negative relationship** between `avg_er` and engagement totals — bigger accounts tend to have lower engagement rates.
- `n_videos` generally scales with engagements, but not consistently (some creators push a lot of content without proportional returns).

## 4. Key Takeaways
- **Multicollinearity** is a major issue if I include all features in a regression-type model.
- **Outliers** and heavy skew could distort the analysis and need to be handled.
- Engagement ratios and `avg_er` provide different insights than raw counts, so they’re worth keeping.

- The **first principal component (PC1)** for both platforms captures the overall scale of engagement (views, likes, shares, etc.), while **PC2** highlights differences in engagement patterns (e.g., relative importance of likes vs. shares).
- PCA scatterplots reveal a wide spread along PC1, confirming the dominance of raw engagement totals, and some separation along PC2, suggesting nuanced differences in how creators drive engagement.

## 2. Regression & GAM Modeling
- Used **Generalized Additive Models (GAMs)** to predict average engagement rate (`avg_er`) from view counts for each platform.
- **Key findings:**
   - There is a clear **negative relationship**: creators with higher view counts tend to have lower engagement rates.
   - The GAM fit is smooth and captures the nonlinear trend, with R² values indicating modest explanatory power.
   - Model predictions (`y_pred_youtube`, `y_pred_tiktok`) closely track the observed engagement rates, but residuals suggest some unmodeled variance.
